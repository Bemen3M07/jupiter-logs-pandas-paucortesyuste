## Pregunta: Què és millor, mostrar els logs a la terminal durant l'execució del programa o bolcar-los en un fitxer de text?

Considerem que és millor bolcar els logs en un fitxer de text. Això es deu al fet que si la terminal es tanca accidentalment, s'executa un `clear` o es perd la sessió, els missatges mostrats a la terminal es perden. En canvi, si els logs es guarden en un fitxer, es poden consultar posteriorment sense perdre cap informació. Això és especialment útil per fer depuració o analitzar esdeveniments passats de manera més ordenada i fiable.

Considero que es mejor guardarlos en un fichero de texto.Porque si la terminal se cierra o se ejecuta un clear se puede perder la sesion, en camvio si se guarda en el fichero se puede conultar cuando se necesario y aprate puedes luego poner comentarion o apuntar la respuesta de dicho error para la proxima vez solucionarlo (Lo que seria un control de errores de toda la vida )


```python
import logging

logger = logging.getLogger("main_logger")
logger.setLevel(logging.DEBUG)

info_handler = logging.FileHandler("info.log")
info_handler.setLevel(logging.INFO)

error_handler = logging.FileHandler("error.log")
error_handler.setLevel(logging.ERROR)

formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
info_handler.setFormatter(formatter)
error_handler.setFormatter(formatter)

logger.addHandler(info_handler)
logger.addHandler(error_handler)

logger.info("info.")
logger.error("error.")

console_handler = logging.StreamHandler()
console_handler.setLevel(logging.INFO)
console_handler.setFormatter(formatter)

logger.addHandler(console_handler)


logger.info("Aquest missatge també es mostra per pantalla.")
logger.error("Aquest error es mostra per pantalla i es guarda a error.log.")
```

# Ejercicio 5 y ejercicio 6


```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {
    'Professor': ['Emili', 'Mario', 'Jaume'],
    'Alumne': ['Xavier', 'Marc', 'Jaume'],
    'M01': [7, 10, 8],
    'M04': [9, 3, 5],
    'M03': [5, 5, 6],
    'M05': [8, 8, 4]
}

df = pd.DataFrame(datos)

print(df)

df['Mitjana'] = df[['M01', 'M04', 'M03', 'M05']].mean(axis=1)
print("\nMitjana de nota final per alumne:")
print(df[['Alumne', 'Mitjana']])

mitjana_total = df['Mitjana'].mean()
print(f"\nMitjana de nota final de tots els alumnes: {mitjana_total:.2f}")

aprovats = df['Mitjana'] >= 5
percentatge_aprovats = (aprovats.sum() / len(df)) * 100
percentatge_suspesos = 100 - percentatge_aprovats
print(f"\nPercentatge de aprovados: {percentatge_aprovats:.2f}%")
print(f"porcentaje de suspesos: {percentatge_suspesos:.2f}%")

notes_mes_baixes = df[['M01', 'M04', 'M03', 'M05']].min()
notes_mes_altes = df[['M01', 'M04', 'M03', 'M05']].max()
print("\nNotes mas bajas:")
print(notes_mes_baixes)
print("\nNotas mas altas")
print(notes_mes_altes)

# Gràfic 1: Evolució de notes per mòduls
plt.figure(figsize=(10, 6))
for columna in ['M01', 'M04', 'M03', 'M05']:
    plt.plot(df['Alumne'], df[columna], marker='o', label=columna)

plt.xlabel('Alumne')
plt.ylabel('Calificación')
plt.title('Calificaciones por Alumno y Módulo')
plt.legend()
plt.show()

# Gràfic 2: Mitjana per alumne
plt.figure(figsize=(10, 6))
plt.bar(df['Alumne'], df['Mitjana'], color='green')
plt.xlabel('Alumne')
plt.ylabel('mediana de la Nota')
plt.title('mediana de la  Nota Final por alumno')
plt.show()

# Gràfic 3: Percentatge d'aprovats i suspesos
plt.figure(figsize=(8, 8))
plt.pie([percentatge_aprovats, percentatge_suspesos], labels=['Aprovats', 'Suspesos'], autopct='%1.1f%%', startangle=90, colors=['blue', 'red'])
plt.title('Porcentaje de aprovados y suspendidos)
plt.show()

# Gràfic 4: Distribució de notes per mòduls (apilats)
plt.figure(figsize=(10, 6))
bottom = [0] * len(df['Alumne'])
for columna in ['M01', 'M04', 'M03', 'M05']:
    plt.bar(df['Alumne'], df[columna], bottom=bottom, label=columna)
    bottom = [bottom[i] + df[columna].iloc[i] for i in range(len(bottom))]

plt.xlabel('Alumne')
plt.ylabel('Notas por modulo')
plt.title('Notas por modulos y alumnos (Apilats)')
plt.legend(title="Mòduls")
plt.show()
```
